```python
from nltk import word_tokenize, sent_tokenize, Text, pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from collections import defaultdict
import cPickle as pickle
import re, random
```

```python
folder = "TAwebpages/Dolphin_Bay"
filename = "%s/reviewtext.pyvar" %folder
filename = open(filename, 'r')
raw_reviews = pickle.load(filename)
filename.close()
```

```python
reviewids = raw_reviews.keys()
len(reviewids)
```

```python
sent_tokens_byreviewid = {}
for reviewid in reviewids:
    rawtext = raw_reviews[reviewid]
    rawtext = rawtext.replace("/", " ")
    rawtext = str(unicode(rawtext, 'ascii', 'ignore'))
    rawtext = rawtext.lower()
    sent_tokens = sent_tokenize(rawtext)
    sent_tokens_byreviewid[reviewid] = sent_tokens
```

```python
word_tokens_byreviewid = {}
for reviewid in reviewids:
    sent_tokens = sent_tokens_byreviewid[reviewid]
    word_tokens_byreviewid[reviewid] = []
    for sent in sent_tokens:
        word_tokens = word_tokenize(sent)
        word_tokens_byreviewid[reviewid].append(word_tokens)
```

```python
postags_byreviewid = {}
for reviewid in reviewids:
    postags_byreviewid[reviewid] = []
    for word_tokens in word_tokens_byreviewid[reviewid]:
        postags = pos_tag(word_tokens)
        postags_byreviewid[reviewid].append(postags)
```

```python
wnl = WordNetLemmatizer()
nouns = set(["NN", "NNS", "NNP", "NNPS"])
ignore = set(stopwords.words('english'))
all_noun_tokens = []
for reviewid in reviewids:
    sents = postags_byreviewid[reviewid]
    for sent in sents:
        for word, pos in sent:
            if ((pos in nouns) and 
                (word not in ignore) and
                (word.isalpha())):
                all_noun_tokens.append(wnl.lemmatize(word)) 
all_noun_tokens = Text(all_noun_tokens)
```

```python
noun_document_count = defaultdict(int)
for term in set(all_noun_tokens):
    for reviewid in reviewids:
        done = False
        for sent in postags_byreviewid[reviewid]:
            if not done:
                for w, pos in sent:
                    if (pos in nouns) and ((w == term) or wnl.lemmatize(w) == term):
                        noun_document_count[term] += 1
                        done = True
```

```python
for noun in noun_document_count:
    count = noun_document_count[noun]
    if count >= len(reviewids)*0.20:
        print count, "\t", noun
```
